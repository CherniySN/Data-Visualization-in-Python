```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('wage-data-coast.csv')
df.head()
```

```python
data = df[df['State'] == 'California'][['Year', 'Salary']]
```

```python
plt.plot(data)
plt.title('Миниммальная з\п в штате Калифорния')
plt.xlabel('Год')
plt.ylabel('Минимальная зарплата $ в час')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,8))
ax.plot(data['Year'],data['Salary'])
ax.set_title('podpis', pad = 16, color = 'Navy', backgroundcolor = 'Gray')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_xticks(list(range(data['Year'].min(), data['Year'].max(), 10))+ [data['Year'].max()])
plt.show()
```

Задание 1
Используя объектно-ориентированный подход, постройте линейный график минимальной зарплаты в штате Аризона до 2000 года включительно. После этого отформатируйте график, сделав заголовок и подписи осей темно-синими.

```python
data = df[(df['State'] == "Arizona") & (df['Year'] <= 2000)][['Salary' , 'Year']]
data1 = df[(df['State'] == 'California') & (df['Year'] <= 2000)][['Salary', 'Year']]

fig, ax= plt.subplots(figsize = (12,8))
ax.axvline(x = 1976, color = 'Red', linewidth = 1, linestyle = '--')
ax.plot(data['Year'], data['Salary'], linewidth = 2.0, label = 'Arizona')
ax.plot(data1['Year'], data1['Salary'], linewidth = 2.0, label = 'California')
ax.set_title('Минимальная зарплата в штате Аризона до 2000 года', pad =16, color = 'Green', backgroundcolor = 'Yellow')
ax.set_xlabel('Год', color = 'Green', backgroundcolor = 'Yellow')
ax.set_ylabel('$ / час', color = 'Green', backgroundcolor = 'Yellow')
ax.set_xticks(list(range(data['Year'].min(), data['Year'].max(), 1)) + [data['Year'].max()])
plt.setp(ax.get_xticklabels(), rotation = 45)
ax.legend()
plt.show()
```

Задание 1
Сравните распределения минимальных зарплат в 1980, 1990 и 2000 годах, построив 3 графика на одной координатной сетке. При постройке графиков на одной координатной сетке убедитесь, что столбцы всех трех графиков имеют одинаковую ширину столбцов, и все 3 графика одновременно видно.

```python
data = df[df['Year'] == 1980]['Salary']
data1 = df[df['Year'] == 1990]['Salary']
data2 = df[df['Year'] == 2000]['Salary']

fig, ax = plt.subplots(figsize = (12,6))

_, bins,_=ax.hist(data2, label = '2000', alpha = 0.5, bins =20)
ax.hist(data, label = '1980', alpha = 0.5, bins = bins)
ax.hist(data1, label = '1990', alpha = 0.5, bins = bins)

ax.set_title('Распределения минимальных зарплат в 1980, 1990 и 2000 годах')
ax.set_ylabel('# записей')
ax.set_xlabel('Минимальная зарплата $ / час')

plt.setp(ax.get_xticklabels(), rotation = 45)
ax.legend()

plt.show()
```

```python
fig, ax = plt.subplots(figsize = (12,6))

for year in [1980,2000, 2010]:
    data = df[df['Year'] == year]['Salary']
    _, bins,_=ax.hist(data, label = year, alpha = 0.5)
    
    ax.set_title('Распределения минимальных зарплат в 1980, 1990 и 2000 годах')
    ax.set_ylabel('# записей')
    ax.set_xlabel('Минимальная зарплата $ / час')

    plt.setp(ax.get_xticklabels(), rotation = 45)
    ax.legend()

plt.show()
```

```python
for year in [1980,2000, 2010]:
    fig, ax = plt.subplots(figsize = (12,6))
    data = df[df['Year'] == year]['Salary']
    _, bins,_=ax.hist(data, label = year, bins =20)
    
    ax.set_title('Распределения минимальных зарплат в' + str(year) + 'годах')
    ax.set_ylabel('# записей')
    ax.set_xlabel('Минимальная зарплата $ / час')

    plt.setp(ax.get_xticklabels(), rotation = 45)
    ax.legend()

plt.show()
```

```python
fig, ax =plt.subplots(nrows=3, ncols=1,figsize= (15,15))
State =['California', 'Alaska', 'Washington']

for index,state in enumerate(State):
    data = df[df['State'] == state][['Salary','Year']]
    ax[index].plot(data['Year'], data['Salary'], linewidth = 2.0, label = str(state))
    ax[index].set_title('Минимальная зарплата в шатате ' + str(state), y = 0.9)
    ax[index].set_xlabel('Года')
    ax[index].set_ylabel('$ / час')
    ax[index].set_xticks(list(range(data['Year'].min(), data['Year'].max(), 4)) + [data['Year'].max()])
    ax[index].legend()
    
plt.show()
```
